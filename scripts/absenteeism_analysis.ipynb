## we aim to reduce the dimensions of the columns by  grouping together similar illnesses

```python
reason_columns.loc[:, "1":"14"]
```

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis="columns")
reason_type_2 = reason_columns.loc[:,15:17].max(axis="columns")
reason_type_3 = reason_columns.loc[:,18 :21].max(axis="columns")
reason_type_3 = reason_columns.loc[:,22 :].max(axis="columns")

reason_type_1.head()
```

## we join the categorized values of the dummy_variables created from  dummy variables and the df column where reason and index has been  dropped

```python
df = pd.concat([df,reason_type_1, reason_type_2, reason_type_3, reason_type_1] , axis ="columns")
```
